```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load the data
df = pd.read_csv('../data/bitcoin_historical_data.csv')  # Adjust filename if needed

df.head()
```

```python
# Convert Timestamp to datetime
df['Timestamp'] = pd.to_datetime(df['Timestamp'], unit='s')  # Convert from Unix timestamp

# Explore
print("Dataset Shape:", df.shape)
print("\nFirst 5 rows:")
print(df.head())
print("\nData Types:")
print(df.info())
print("\nBasic Statistics:")
print(df.describe())

# Check for missing values
print("\nMissing Values:")
print(df.isnull().sum())

# Simple visualization: Price over time
plt.figure(figsize=(14, 6))
plt.plot(df['Timestamp'], df['Close'], linewidth=1, color='blue')
plt.title('Bitcoin Price Over Time', fontsize=14, fontweight='bold')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(df['Close'], bins=50, color='green', alpha=0.7, edgecolor='black')
plt.title('Bitcoin Price Distribution', fontsize=14, fontweight='bold')
plt.xlabel('Price (USD)')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
# Volume over time (simplified)
plt.figure(figsize=(14, 6))
plt.fill_between(df['Timestamp'], df['Volume'], color='steelblue', alpha=0.5)
plt.title('Bitcoin Trading Volume Over Time', fontsize=14, fontweight='bold')
plt.xlabel('Date')
plt.ylabel('Volume')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
